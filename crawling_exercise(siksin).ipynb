```python
import pandas as pd
from bs4 import BeautifulSoup
from urllib.parse import quote
import requests
```

# 1. 데이터 가져오기

```python
quote('신논현역')
```

```python
site = 'https://www.siksinhot.com/ko'
url = f'{site}/search?keywords={quote("신논현역")}'
req = requests.get(url)
html = req.text
```

```python
html
```

```python
soup = BeautifulSoup(req.text, 'html.parser')
```

```python
soup
```

# 2. 원하는 데이터 찾기

```python
lis = soup.find_all('li')
len(lis)
```

```python
lis = soup.select_one('.listTy1').find_all('li')
lis
```

```python
lis = soup.select('.listTy1 > ul > li') #>자식 tag
lis
```

3. 하나를 선택해서 원하는 정보를 추출

상호, 평점, 메뉴

```python
li = lis[0]
li
```

```python
store = li.select_one('.box_tit').get_text()
store
```

```python
#float(), int()로 변환하고자 할 때는 반드시 예외처리를 해주어야함
try :
  score = float(li.select_one('.score').get_text())
except:
  score = 0.0

score
```

```python
place = li.select_one('.cnt > ul').get_text()
place                                       
```

```python
menu = li.select_one('.cnt > p').get_text()
menu
```

# 4. 반복문을 사용해서 페이지에 있는 데이터를 가져오기

```python
line=[]
for li in lis:
  store=li.select_one(' .box_tit').get_text()
  try:
    score= float(li.select_one('.score').get_text())
  except:
    score=0.0
  place = li.select_one('.cnt > ul').get_text()
  menu=li.select_one('.cnt > p').get_text()
  line.append({'상호':store,'평점':score,'위치':place,'메뉴':menu})
line
```

```python
df=pd.DataFrame(line)
df
```

```python
df.to_csv('식신맛집(신논현역).csv', index=False)
```
